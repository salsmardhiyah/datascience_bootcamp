### 3. Mengolah Dataset (30pts)

Dengan menggunakan dataset `application_processed.csv` buatlah dataframe baru dengan ketentuan:
1. Ambil semua users yang memiliki `email`. (ditandai dengan nilai pada column `flag_email` = 1) dan tidak memilik data `duplicates`. (5pts)
2. Buatlah column baru bernama `has_car_and_property` dimana nilainya merupakan range dari `0-1`. `1` apabila `flag_own_car` bernilai `1` dan `flag_own_property` bernilai `1`. `0` apabila terdapat atau semua nilai pada column `flag_own_car` dan `flag_own_property` terdapat angka 0. (5pts)
3. Ambil column yang diperlukan yaitu `id`, `gender`, `has_car_and_property`, `income_type`, `education_type`, `family_status`, `housing_type`, `occupation_type`, `amount_income_total`. (5pts)
4. Lakukan filter dengan ketentuan sebagai berikut: (5pts)
    - Memiliki `car` dan `property`
    - Tidak ada nilai null pada column `occupation_type`
    - Memiliki value pada column `occupation_type` dengan pola `staff`.
5. Sudah dilakukan sorting berdasarkan `income_type` dari `A-Z` dan `amount_income_total` dari `besar ke kecil`. (5pts)
6. Simpan kedalam bentuk `.csv` dengan nama `homework_3_result.csv`. (5pts)

**Expected Result:**

![Expected Result](assets/homework_exercise_3_result.png)

```python
## Your code below here
#1. Ambil semua users yang memiliki email. (ditandai dengan nilai pada column flag_email = 1) dan tidak memilik data duplicates. (5pts)

#cek apakah terdapat data duplikat
application.duplicated().any() #True

#melihat data duplikat
application[application.duplicated(keep=False) == True] 

#menghilangkan data duplikat
application_clean = application.drop_duplicates() 

#cek kembali apakah terdapat data duplikat
application_clean.duplicated().any() #False

#mengambil semua user yang memiliki email
email_users = application_clean[application_clean['flag_email'] == 1] 
email_users.head()

```

```python
#2. Buatlah column baru bernama has_car_and_property dimana nilainya merupakan range dari 0-1. 
#1 apabila flag_own_car bernilai 1 dan flag_own_property bernilai 1. 
#0 apabila terdapat atau semua nilai pada column flag_own_car dan flag_own_property terdapat angka 0. (5pts)

# membuat column baru dengan constraint
application_new = email_users.assign(has_car_and_property = (email_users['flag_own_car'] == 1)
            & (email_users['flag_own_property'] == 1))

#mengubah tipe data boolean menjadi int 0-1
application_new['has_car_and_property'] = application_new['has_car_and_property'].astype(int)
application_new.head()

```

```python
#3. Ambil column yang diperlukan yaitu id, gender, has_car_and_property, income_type, education_type, family_status, housing_type, occupation_type, amount_income_total. (5pts)
application_column = application_new[['id',
                                      'gender',
                                      'has_car_and_property', 
                                      'income_type', 
                                      'education_type', 
                                      'family_status', 
                                      'housing_type', 
                                      'occupation_type', 
                                      'amount_income_total'
                                     ]]
application_column.head()
```

```python
#4. Lakukan filter dengan ketentuan sebagai berikut: (5pts)
#Memiliki car dan property
#Tidak ada nilai null pada column occupation_type
#Memiliki value pada column occupation_type dengan pola staff.

application_column_filter = application_column[(application_column['has_car_and_property'] == 1)
                                               & (application_column['occupation_type'].str.contains('staff$')) 
                                              ] #baris yang memiliki occupation_type null otomatis tidak valid
application_column_filter.head()
```

```python
#5. Sudah dilakukan sorting berdasarkan income_type dari A-Z dan amount_income_total dari besar ke kecil
application_sorted = application_column_filter.sort_values(['income_type', 'amount_income_total'], ascending=[True, False])

#melakukan reset index
application_reset_index = application_sorted.reset_index(drop=True)
application_reset_index.head()
```

```python
#6. Simpan kedalam bentuk .csv dengan nama homework_3_result.csv. (5pts)
application_reset_index.to_csv("homework_3_result.csv", index=None)

#melihat data file homework_3_result.csv
pd.read_csv('homework_3_result.csv')
```

### 5. Memanfaatkan Pivot Tabel (15pts)

Buatlah dataframe baru menggunakan fungsi pivot table untuk melihat `Embarked` apa saja yang `average_female_ticket_price` lebih besar daripada `average_male_ticket_price`. Informasi `average_female_ticket_price` dan `average_male_ticket_price` didapatkan dari rata-rata pada column `Fare`.

**Hint:**
1. menggunakan pivot table
2. dapat dilihat pada topik `DataFrame Aggregation` bagian `Pivot Table | Reshape Columns and Rows`

**Expected Result:**

![Expected Result](assets/homework_exercise_5_result.png)

```python
## Your code below here

#melakukan pivot table dan reset index
titanic_pivot = pd.pivot_table(titanic,
                               index = ['Embarked'],
                               columns = ['Sex'],
                               values = ['Fare'],
                               aggfunc = 'mean').reset_index()


#merubah nama column
titanic_pivot.columns = ['Embarked', 'avg_female_ticket_price', 'avg_male_ticket_price']

#melakukan filter avg_female_ticket_price > avg_male_ticket_price
titanic_pivot[titanic_pivot['avg_female_ticket_price'] > titanic_pivot['avg_male_ticket_price']]
```

### 7. Merge DataFrame (25pts)

Buatlah 2 dataframe mengikuti petunjuk berikut:

df_1 :  Filter dataset 01 telecom_revenue.csv dengan multiple kondisi (5pts)
- MonthlyRevenue > 10 
- Occupation terdiri dari Professional, Student, and Crafts

df_2 :  Filter dataset 02 telecom_usage.csv dengan multiple kondisi (5pt)
- nilai UnansweredCalls > BlockedCalls

Jawablah pertanyaan berikut ini:
1. Hitunglah total CustomerID dan rata-rata DroppedCalls untuk masing-masing occupation. (10 Points)
2. Pada occupation apa, rata-rata DroppedCalls paling besar? (5 Points)

**Expected Result:**

![Expected Result](assets/homework_exercise_7_result.png)

```python
## Your code below here
#membaca file 01 telecom_revenue.csv
rev = pd.read_csv('/Users/salsa/Data Science/DS-31/Week 5/Homework Data Processing/01 telecom_revenue.csv')

# MonthlyRevenue > 10
# Occupation terdiri dari Professional, Student, and Crafts
df_1 = rev[
            (rev['MonthlyRevenue'] > 10) &
            (rev['Occupation'].isin(['Professional', 'Student', 'Crafts']))
            ]
df_1.head()
```

```python
#membaca file 02 telecom_usage.csv
usage = pd.read_csv('/Users/salsa/Data Science/DS-31/Week 5/Homework Data Processing/02 telecom_usage.csv')
usage.head()

# nilai UnansweredCalls > BlockedCalls
df_2 = usage[
            usage['UnansweredCalls'] > usage['BlockedCalls']
            ]

df_2.head()
```

```python
#melakukan merge pada dua data frame
df_merged = df_1.merge(df_2, how = 'inner', on = 'CustomerID')

#1. Hitunglah total CustomerID dan rata-rata DroppedCalls untuk masing-masing occupation. (10 Points)
df_merged_occupation = df_merged.groupby(df_merged['Occupation']).agg({'CustomerID' : ['nunique'],
                                                                       'DroppedCalls' : ['mean']
                                                                      })
df_merged_occupation
```

```python
#2. Pada occupation apa, rata-rata DroppedCalls paling besar? (5 Points)

print(df_merged_occupation['DroppedCalls'].idxmax())

print('rata-rata DroppedCalls paling besar yaitu pada Occupation Student')
```

```python
def string_processor(kata, proses):
    ''' (str, str) -> str / int
    function string_processor dengan inputan 2 parameter, 
    yaitu suatu string 1 kata dan nama proses yang akan dilakukan

    Function ini memiliki 4 proses:
    first_char : return huruf pertama dari string yang diinput
    last_char : return huruf terakhir dari string yang diinput
    num_chars : return jumlah huruf string input
    num_vowels : return jumlah huruf vocal pada string input
    '''

    proses = proses.lower()
    vowels = 0

    if proses == 'first_char':
        return kata[0]
    elif proses == 'last_char':
        return kata[-1]
    elif proses == 'num_chars':
        return len(kata)
    elif proses == 'num_vowels':
        for huruf in kata:
            if huruf in 'aAiIuUeEoO':
                vowels = vowels + 1
        return vowels
    else:
        return 'silahkan masukkan parameter proses dengan benar: first_char/last_char/num_chars/num_vowels'
```

```python
#  melakukan import data xlsx dari google drive
id2 = '15UQtE4rvJE_uyD0tYh4aVJTnfyhC0YZG' #id file di google drive

import pandas as pd

downloaded = drive.CreateFile({'id':id2}) 
downloaded.GetContentFile('superstore.xls')  
df_superstore = pd.read_excel('superstore.xls')
df_superstore.head(2)
```

```python
# cek data duplikat
df_superstore.duplicated().any()
```

```python
# menambahkan kolom order_year ke dalam data frame
df_superstore['order_year'] = df_superstore['Order Date'].apply(lambda x: str(x.year))
df_superstore.head(3)
```

```python

# membuat function 
def get_sales_profit(region, year='2017'):
    ''' (str, str) -> data frame
    
    Function ini dapat menghitung Jumlah Sales dan Profit pada suatu tahun dan region tertentu dengan default tahun 2017. 
    '''

    # filter dataframe berdasarkan region dan year
    df_by_region_year = df_superstore[(df_superstore.Region == region) & \
                                        (df_superstore.order_year == year)]

    # mendapatkan jumlah profit dan sales 
    df_total = df_by_region_year.groupby(['order_year','Region']).agg({'Sales' : ['sum'], 
                                                            'Profit' : ['sum']})
    
    #merubah nama kolom
    df_total.columns = ['total_sales', 'total_profit']

    # function return
    return df_total
    
```

```python
# menentukan sales dan profit South Region pada tahun 2016
get_sales_profit('South', '2016')
```

```python
# menentukan sales dan profit West Region pada tahun 2017 dengan parameter default
get_sales_profit('West')
```

```python
# menentukan sales dan profit West Region pada tahun 2017
get_sales_profit('West', '2017')
```

```python
#  melakukan import data xls dari google drive
id2 = '15UQtE4rvJE_uyD0tYh4aVJTnfyhC0YZG' #id file di google drive

import pandas as pd

downloaded = drive.CreateFile({'id':id2}) 
downloaded.GetContentFile('superstore.xls')  
df7 = pd.read_excel('superstore.xls')
df7.head(2)
```

```python
# cek data duplikat
df7.duplicated().any()
```

```python
# menambahkan kolom year ke dalam data frame
df7['year'] = df7['Order Date'].apply(lambda x: str(x.year))
df7.head(2)
```

```python
# melakukan agregasi total sales untuk tiap tahun dan customer id
agregat = df7.groupby(['year', 'Customer ID']).agg({'Sales' : 'sum'}).reset_index()
agregat.sample(2)
```

```python
# melihat data statistik deskriptif pada sales pertahun
statistik = agregat.groupby('year')['Sales'].describe().reset_index()
statistik
```

```python
# mengambil kolom yang diperlukan
statistik = statistik[['year', '25%', '50%', '75%']]
statistik
```

```python
# mengubah nama kolom
statistik.columns = ['year', 'p-25', 'p-50', 'p-75']
statistik
```

```python
# melakukan merging data frame statistik dan agregat
merged_data = agregat.merge(statistik, on = 'year')
merged_data.sample()
```

```python
# membuat fungsi segmentasi sales performance customer dengan exception handling
def get_segment(df):
    try:
        if df['Sales'] > df['p-75']:
            return 'High'
        elif df['Sales'] >= df['p-50']:
            return 'Medium'
        else:
            return 'Low'
    
    except Exception as e:
        print('Error', e)
```

**Alasan dalam penentuan kriteria segmentasi customer yang dibuat:**
* nilai persentil 50 dan 75 tiap tahunnya mengalami kenaikan yang stabil
* nilai persentil 25 mengalami kenaikan yang fluktuatif tiap tahunnya
* sehingga persentil 50 dan 75 dapat dimanfaatkan sebagai batas segmentasi sales performance customer

```python
# membuat kolom baru segment
merged_data['Segment'] = merged_data.apply(lambda x: get_segment(x), axis = 1)
merged_data.head(2)
```

```python
# menentukan jumlah customer pada masing-masing segment di tahun 2017
jumlah_per_segment = merged_data[merged_data['year'] == '2017'].groupby(['year', 'Segment']).agg({'Customer ID' : 'nunique'})

# merubah nama kolom
jumlah_per_segment.columns = ['jumlah_customer']
jumlah_per_segment
```
